## Encoding

```python
gender = {'Male': 1,
             'Female': 0}
df['Gender'] = df['Gender'].map(gender)


married = {'Yes': 1,
             'No': 0}
df['Married'] = df['Married'].map(married)


dependents = {'0': 0,
             '1': 1,
             '2': 2,
             '3+' : 3}
df['Dependents'] = df['Dependents'].map(dependents)


education = {'Graduate': 1,
             'Not Graduate': 0}
df['Education'] = df['Education'].map(education)


self_Employed = {'Yes': 1,
             'No': 0}
df['Self_Employed'] = df['Self_Employed'].map(self_Employed)


property_Area = {'Urban' : 2,
             'Semiurban': 1,
             'Rural': 0}
df['Property_Area'] = df['Property_Area'].map(property_Area)


loan_Status = {'Y': 1,
             'N': 0}
df['Loan_Status'] = df['Loan_Status'].map(loan_Status)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

## Standerdization & Logistic Regression

```python
x = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(x_train_scaled, y_train)   
```

```python
y_pred = logreg.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report")
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob = logreg.predict_proba(x_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False +ve Rate")
plt.ylabel("True +ve Rate")
plt.show()

auc_score = roc_auc_score(y_test, y_prob)
print(f"AUC Score: {auc_score}")
```

## Regularization

```python
# L2 Regularization (Ridge)
lr_l2 = LogisticRegression(
    penalty='l2', 
    C=0.5, 
    solver='liblinear', 
    random_state=42
)

lr_l2.fit(x_train_scaled, y_train)
y_pred_l2 = lr_l2.predict(x_test_scaled)

print("L2 Regularization Results")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_l2))
print("\nClassification Report (L2):\n", classification_report(y_test, y_pred_l2))
print("\nNumber of ZERO coefficients (L2):", np.sum(lr_l2.coef_ == 0))
```

```python
cm = confusion_matrix(y_test, y_pred_l2)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob = lr_l2.predict_proba(x_test_scaled)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False +ve Rate")
plt.ylabel("True +ve Rate")
plt.show()

auc_score = roc_auc_score(y_test, y_prob)
print(f"AUC Score: {auc_score}")
```

```python
# L1 Regularization (Lasso)
lr_l1 = LogisticRegression(
    penalty='l1', 
    C=0.5, 
    solver='liblinear', 
    random_state=42
)

lr_l1.fit(x_train_scaled, y_train)
y_pred_l1 = lr_l1.predict(x_test_scaled)

print("Regularization Results")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_l1))
print("\nClassification Report (L1):\n", classification_report(y_test, y_pred_l1))
print("\nNumber of ZERO coefficients (L1):", np.sum(lr_l1.coef_ == 0))
```

```python
cm = confusion_matrix(y_test, y_pred_l1)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob = lr_l1.predict_proba(x_test_scaled)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False +ve Rate")
plt.ylabel("True +ve Rate")
plt.show()

auc_score = roc_auc_score(y_test, y_prob)
print(f"AUC Score: {auc_score}")
```

## Logistic Regression with Grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Base Logistic Regression model
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Hyperparameter grid
param_grid = [
    {
        'penalty': ['l1', 'l2', 'elasticnet', 'none'],
        'C': [0.01, 0.1, 0.5, 1, 5, 10],
        'solver': ['saga'],   # saga supports all penalties including elasticnet
        'l1_ratio': [0, 0.25, 0.5, 0.75, 1]  # only used for elasticnet
    }
]
```

```python
# GridSearch with 5-fold cross-validation
grid_logreg = GridSearchCV(
    estimator=logreg,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit GridSearch
grid_logreg.fit(x_train_scaled, y_train)

# Best hyperparameters
print("Best Parameters:", grid_logreg.best_params_)
print("Best CV Score:", grid_logreg.best_score_)
```

```python
# Predict using best estimator
best_logreg = grid_logreg.best_estimator_
y_pred_best = best_logreg.predict(x_test_scaled)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

print("=== Tuned Logistic Regression Evaluation ===")
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print("Precision:", precision_score(y_test, y_pred_best))
print("Recall:", recall_score(y_test, y_pred_best))
print("F1 Score:", f1_score(y_test, y_pred_best))

print("\nClassification Report:\n", classification_report(y_test, y_pred_best))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Predicted probabilities
y_prob_best = best_logreg.predict_proba(x_test_scaled)[:,1]

fpr, tpr, _ = roc_curve(y_test, y_prob_best)
auc_score = roc_auc_score(y_test, y_prob_best)

plt.plot(fpr, tpr, marker='.')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"Tuned Logistic Regression ROC Curve (AUC = {auc_score:.2f})")
plt.show()

print(f"AUC Score: {auc_score}")
```

## Decision Tree

### Gini

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

dt_gini = DecisionTreeClassifier(
    criterion="gini",
    max_depth=5,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)

dt_gini.fit(x_train_scaled, y_train)
y_pred_gini = dt_gini.predict(x_test_scaled)

print("=== Decision Tree (Gini) ===")
print("Accuracy:", accuracy_score(y_test, y_pred_gini))
print("Precision:", precision_score(y_test, y_pred_gini))
print("Recall:", recall_score(y_test, y_pred_gini))
print("F1 Score:", f1_score(y_test, y_pred_gini))
print("\nClassification Report:\n", classification_report(y_test, y_pred_gini))
```

### Entropy

```python
dt_entropy = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=5,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)

dt_entropy.fit(x_train_scaled, y_train)
y_pred_entropy = dt_entropy.predict(x_test_scaled)

print("=== Decision Tree (Entropy) ===")
print("Accuracy:", accuracy_score(y_test, y_pred_entropy))
print("Precision:", precision_score(y_test, y_pred_entropy))
print("Recall:", recall_score(y_test, y_pred_entropy))
print("F1 Score:", f1_score(y_test, y_pred_entropy))
print("\nClassification Report:\n", classification_report(y_test, y_pred_entropy))
```

```python
print("\n=== Model Comparison ===")
print("Gini Accuracy:", accuracy_score(y_test, y_pred_gini))
print("Entropy Accuracy:", accuracy_score(y_test, y_pred_entropy))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))

plot_tree(
    dt_gini, 
    feature_names=x.columns,
    class_names=["Not Approved", "Approved"],
    filled=True,
    rounded=True,
    fontsize=9
)

plt.title("Decision Tree (Gini)")
plt.show()
```

```python
importance = pd.DataFrame({
    "Feature": x.columns,
    "Importance": dt_gini.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\nFeature Importance (Decision Tree - Gini):")
print(importance)
```
